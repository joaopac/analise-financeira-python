# Aplicação Prática de Python na Ciência dos Dados
## Explorando as Características Financeiras do Dataset

Neste notebook, analisamos as características financeiras do dataset `default_of_credit_card_clients`, explorando variáveis de valor da fatura (BILL_AMT) e valor dos pagamentos (PAY_AMT). Utilizaremos bibliotecas como **Pandas**, **NumPy** e **Matplotlib** para manipulação e visualização dos dados.

```python
# Importação das bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Carregamento do dataset
df = pd.read_excel('default_of_credit_card_clients__courseware_version_1_21_19.xls', header=1)

print('Dataset carregado com sucesso!')
print(f'Shape: {df.shape}')
df.head()
```

---
## Exercício 1 (2,0 pts)
### Criar listas com nomes para as características financeiras restantes

Criamos duas listas separadas: uma para as características de **valor da fatura** (BILL_AMT) e outra para as características de **valor do pagamento** (PAY_AMT), organizando assim as variáveis financeiras do dataset.

```python
# Lista com as características de valor da fatura
bill_amt_features = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']

# Lista com as características de valor do pagamento
pay_amt_features = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

print('Características de valor da fatura (BILL_AMT):')
print(bill_amt_features)

print('\nCaracterísticas de valor do pagamento (PAY_AMT):')
print(pay_amt_features)
```

---
## Exercício 2 (3,0 pts)
### Resumo estatístico das características de valor da fatura (BILL_AMT)

Utilizamos o método `.describe()` para examinar as sínteses estatísticas das variáveis BILL_AMT. Esse método retorna métricas como média, desvio padrão, valores mínimo e máximo, e os quartis.

```python
# Resumo estatístico das características de valor da fatura
bill_describe = df[bill_amt_features].describe()
print('Resumo estatístico - Valor da Fatura (BILL_AMT):')
bill_describe
```

### Reflexão sobre os resultados

Os resultados do `.describe()` fazem sentido para dados de faturas de cartão de crédito:

- **Valores mínimos negativos**: É possível que clientes tenham crédito na fatura (ex: estorno ou pagamento a maior), o que justifica valores negativos em BILL_AMT.
- **Média e mediana distantes**: A diferença entre a média e a mediana (50%) indica que a distribuição é assimétrica — há clientes com faturas muito altas que puxam a média para cima.
- **Desvio padrão elevado**: Confirma grande variabilidade nos valores de fatura entre os clientes.
- **Valores máximos muito altos**: Indicam a presença de clientes com limites e gastos elevados, o que é esperado em datasets de cartão de crédito.

Portanto, os dados fazem sentido dentro do contexto de um dataset financeiro de cartão de crédito.

---
## Exercício 3 (4,0 pts)
### Visualização das características de valor da fatura em grade 2x3 de histogramas

Visualizamos a distribuição das variáveis BILL_AMT usando uma grade de 2 linhas por 3 colunas de histogramas, com 20 bins cada.

```python
# Grade 2x3 de histogramas para as características de valor da fatura
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
fig.suptitle('Distribuição das Características de Valor da Fatura (BILL_AMT)', fontsize=14, fontweight='bold')

for i, feature in enumerate(bill_amt_features):
    row = i // 3
    col = i % 3
    axes[row, col].hist(df[feature], bins=20, color='steelblue', edgecolor='white')
    axes[row, col].set_title(feature, fontsize=11)
    axes[row, col].set_xlabel('Valor (NT$)')
    axes[row, col].set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

---
## Exercício 4 (4,0 pts)
### Resumo estatístico das características de valor do pagamento (PAY_AMT)

Agora aplicamos o `.describe()` às variáveis PAY_AMT para entender a distribuição dos valores de pagamento realizados pelos clientes.

```python
# Resumo estatístico das características de valor do pagamento
pay_describe = df[pay_amt_features].describe()
print('Resumo estatístico - Valor do Pagamento (PAY_AMT):')
pay_describe
```

### Reflexão sobre os resultados

Os resultados do `.describe()` para PAY_AMT também fazem sentido:

- **Valor mínimo igual a 0**: Diferente das faturas, os pagamentos não podem ser negativos — o mínimo de 0 indica clientes que não fizeram nenhum pagamento no período.
- **Mediana próxima de 0 ou muito baixa**: Grande parte dos clientes faz pagamentos mínimos ou não paga, o que é coerente com um dataset voltado ao estudo de inadimplência.
- **Média maior que a mediana**: A distribuição é fortemente assimétrica à direita, com alguns clientes realizando pagamentos muito elevados.
- **Valores máximos muito altos**: Clientes que quitam grandes faturas elevam o valor máximo consideravelmente.

Esses padrões são consistentes com o comportamento esperado em dados de pagamento de cartão de crédito.

## Exercício 5 (4,0 pts)
### Histogramas das características de valor do pagamento com rotação nos rótulos do eixo x

Plotamos os histogramas das variáveis PAY_AMT com o argumento `xrot=45` para rotacionar os rótulos do eixo x e evitar sobreposição.

```python
# Histogramas das características de valor do pagamento com rotação no eixo x
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
fig.suptitle('Distribuição das Características de Valor do Pagamento (PAY_AMT)', fontsize=14, fontweight='bold')

for i, feature in enumerate(pay_amt_features):
    row = i // 3
    col = i % 3
    axes[row, col].hist(df[feature], bins=20, color='coral', edgecolor='white')
    axes[row, col].set_title(feature, fontsize=11)
    axes[row, col].set_xlabel('Valor (NT$)')
    axes[row, col].set_ylabel('Frequência')
    axes[row, col].tick_params(axis='x', rotation=45)  # xrot=45

plt.tight_layout()
plt.show()
```

### Considerações sobre os resultados

Os histogramas confirmam o que o `.describe()` sugeriu:

- A grande maioria dos pagamentos está concentrada em valores próximos de zero, evidenciando que muitos clientes pagam pouco ou nada.
- Há uma cauda longa à direita, com poucos clientes realizando pagamentos muito elevados.
- Essa distribuição fortemente assimétrica (skewed right) é típica de dados financeiros e justifica o uso de transformações logarítmicas para melhor visualização.

## Exercício 6 (4,0 pts)
### Máscara booleana para verificar pagamentos iguais a zero

Usamos uma máscara booleana para identificar e contar quantos registros possuem valor de pagamento exatamente igual a zero em cada variável PAY_AMT.

```python
# Máscara booleana: valores de pagamento iguais a zero
mask_zeros = df[pay_amt_features] == 0

# Contagem de zeros por coluna
zeros_count = mask_zeros.sum()
total = len(df)

print('Quantidade de pagamentos iguais a zero:')
print(zeros_count)
print(f'\nTotal de registros no dataset: {total}')
print('\nPercentual de zeros por variável:')
print((zeros_count / total * 100).round(2).astype(str) + '%')
```

### O resultado faz sentido?

Sim, os resultados fazem total sentido dado o histograma anterior:

- Os histogramas mostraram uma enorme concentração de valores próximos de zero, e a máscara booleana confirma que uma grande parcela dos clientes possui pagamentos **exatamente iguais a zero**.
- Isso é coerente com um dataset de inadimplência, onde muitos clientes simplesmente não efetuam nenhum pagamento em determinados meses.
- A presença massiva de zeros justifica a necessidade de filtrá-los antes de aplicar a transformação logarítmica, já que `log(0)` é indefinido matematicamente.

---
## Exercício 7 (4,0 pts)
### Histogramas de transformações logarítmicas dos pagamentos diferentes de zero

Ignoramos os pagamentos iguais a zero usando a máscara criada no exercício anterior, e aplicamos `np.log10()` com o método `.apply()` do Pandas para transformar os dados e plotar histogramas mais informativos.

```python
# Filtrando apenas pagamentos diferentes de zero usando a máscara booleana
# A máscara mask_zeros é True onde o valor é zero, então usamos ~mask_zeros para pegar os não-zeros

fig, axes = plt.subplots(2, 3, figsize=(14, 8))
fig.suptitle('Transformação Logarítmica dos Pagamentos (PAY_AMT ≠ 0)', fontsize=14, fontweight='bold')

for i, feature in enumerate(pay_amt_features):
    row = i // 3
    col = i % 3

    # Filtrando os valores diferentes de zero para a coluna atual
    non_zero_data = df[feature][~mask_zeros[feature]]

    # Aplicando np.log10 com o método .apply()
    log_data = non_zero_data.apply(np.log10)

    axes[row, col].hist(log_data, bins=20, color='seagreen', edgecolor='white')
    axes[row, col].set_title(feature, fontsize=11)
    axes[row, col].set_xlabel('log10(Valor)')
    axes[row, col].set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

### Considerações sobre os resultados

A transformação logarítmica revela padrões que eram invisíveis nos histogramas originais:

- **Distribuição mais simétrica**: Após aplicar `log10`, os dados apresentam uma distribuição muito mais próxima da normal (curva em sino), facilitando análises estatísticas.
- **Concentração entre 2 e 4**: Isso corresponde a pagamentos entre NT\$100 e NT\$10.000 na escala original, que representa a faixa mais comum de pagamentos.
- **Redução do efeito dos outliers**: A transformação logarítmica comprime a cauda direita, tornando os histogramas muito mais informativos e legíveis do que os originais.
- **Padrão consistente entre os meses**: Todas as 6 variáveis PAY_AMT apresentam distribuições similares após a transformação, sugerindo comportamento de pagamento estável ao longo dos meses analisados.

Essa é uma técnica fundamental em ciência de dados para lidar com distribuições assimétricas e tornar os dados mais adequados para modelagem preditiva.
